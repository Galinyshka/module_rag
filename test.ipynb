```python
from qdrant_client import QdrantClient
from qdrant_client.models import Filter, FieldCondition, MatchValue, MatchAny

client = QdrantClient(url="http://localhost:6333")

results = client.scroll(
    collection_name="discipline_chunks",
    scroll_filter=Filter(must=[
        FieldCondition(key="discipline", match=MatchValue(value="Семантические технологии")),
    ]),
    limit=100,
    with_payload=True,
    with_vectors=False,
)

for point in results[0]:
    print(point.payload.get("block_type"), "|", point.payload.get("block_name", "")[:50])
```
